# Step 1.1: Compute Match points from Kaggle Dataset

```python
drive.mount('/content/drive')
filtered_df = pd.read_csv('/content/drive/My Drive/CS 441/Project/epl_filtered_2010_2024.csv')

print("Data loaded successfully!")
print(f"Total rows: {len(filtered_df)}")
print(f"Seasons: {filtered_df['Season'].unique()}")
```

```python
filtered_df['HomeTeamPoints'] = 0
filtered_df['AwayTeamPoints'] = 0

filtered_df.loc[filtered_df['FullTimeResult'] == 'H', 'HomeTeamPoints'] = 3
filtered_df.loc[filtered_df['FullTimeResult'] == 'A', 'AwayTeamPoints'] = 3
filtered_df.loc[filtered_df['FullTimeResult'] == 'D', 'HomeTeamPoints'] = 1
filtered_df.loc[filtered_df['FullTimeResult'] == 'D', 'AwayTeamPoints'] = 1

print("Points calculated from match results")
print("\nSample match results with points:")
print(filtered_df[['Season', 'HomeTeam', 'AwayTeam', 'FullTimeResult', 'HomeTeamPoints', 'AwayTeamPoints']].head(10))
```

```python
home_stats = filtered_df.groupby(['Season', 'HomeTeam']).agg({'HomeTeamPoints': 'sum','FullTimeHomeGoals': 'sum','FullTimeAwayGoals': 'sum','HomeShots': 'sum','HomeShotsOnTarget': 'sum','HomeCorners': 'sum','HomeFouls': 'sum','HomeYellowCards': 'sum','HomeRedCards': 'sum'}).reset_index()
home_stats.columns = ['Season', 'Team', 'HomePoints', 'HomeGoalsScored', 'HomeGoalsConceded', 'HomeShots', 'HomeShotsOnTarget', 'HomeCorners', 'HomeFouls', 'HomeYellowCards', 'HomeRedCards']
print("Home statistics aggregated")
print(f"Home stats rows: {len(home_stats)}")
print("\nSample home stats (2023/24 season):")
print(home_stats[home_stats['Season'] == '2023/24'].head())
```

```python
away_stats = filtered_df.groupby(['Season', 'AwayTeam']).agg({'AwayTeamPoints': 'sum','FullTimeAwayGoals': 'sum','FullTimeHomeGoals': 'sum','AwayShots': 'sum','AwayShotsOnTarget': 'sum','AwayCorners': 'sum','AwayFouls': 'sum','AwayYellowCards': 'sum','AwayRedCards': 'sum'}).reset_index()
away_stats.columns = ['Season', 'Team', 'AwayPoints', 'AwayGoalsScored', 'AwayGoalsConceded', 'AwayShots', 'AwayShotsOnTarget','AwayCorners', 'AwayFouls', 'AwayYellowCards', 'AwayRedCards']
print("Away statistics aggregated")
print(f"Away stats rows: {len(away_stats)}")
print("\nSample away stats (2023/24 season):")
print(away_stats[away_stats['Season'] == '2023/24'].head())
```

```python
season_stats = pd.merge(home_stats, away_stats, on=['Season', 'Team'], how='outer').fillna(0)
for col in season_stats.columns:
    if col not in ['Season', 'Team']:
        season_stats[col] = season_stats[col].astype(int)

print("Home and away stats merged")
print("\nSample merged data (2023/24 - Man City):")
man_city_2024 = season_stats[(season_stats['Season'] == '2023/24') & (season_stats['Team'] == 'Man City')]
print(man_city_2024)
```

```python
season_stats['TotalPoints'] = season_stats['HomePoints'] + season_stats['AwayPoints']
season_stats['TotalGoalsScored'] = season_stats['HomeGoalsScored'] + season_stats['AwayGoalsScored']
season_stats['TotalGoalsConceded'] = season_stats['HomeGoalsConceded'] + season_stats['AwayGoalsConceded']
season_stats['GoalDifference'] = season_stats['TotalGoalsScored'] - season_stats['TotalGoalsConceded']
season_stats['TotalShots'] = season_stats['HomeShots'] + season_stats['AwayShots']
season_stats['TotalShotsOnTarget'] = season_stats['HomeShotsOnTarget'] + season_stats['AwayShotsOnTarget']
season_stats['TotalCorners'] = season_stats['HomeCorners'] + season_stats['AwayCorners']
season_stats['TotalFouls'] = season_stats['HomeFouls'] + season_stats['AwayFouls']
season_stats['TotalYellowCards'] = season_stats['HomeYellowCards'] + season_stats['AwayYellowCards']
season_stats['TotalRedCards'] = season_stats['HomeRedCards'] + season_stats['AwayRedCards']

print("\nSample data with all metrics (2023/24 - Top 5 teams by points):")
top_2023 = season_stats[season_stats['Season'] == '2023/24'].nlargest(5, 'TotalPoints')
print(top_2023[['Season', 'Team', 'TotalPoints', 'TotalGoalsScored', 'TotalGoalsConceded', 'GoalDifference']])
```

```python
kaggle_computed_table = season_stats[['Season','Team','TotalPoints','HomePoints','AwayPoints','TotalGoalsScored','HomeGoalsScored','AwayGoalsScored','TotalGoalsConceded','HomeGoalsConceded','AwayGoalsConceded','GoalDifference','TotalShots','TotalShotsOnTarget','TotalCorners','TotalFouls','TotalYellowCards','TotalRedCards']]
kaggle_computed_table = kaggle_computed_table.sort_values(['Season', 'TotalPoints'], ascending=[True, False])
print(f"\nTotal rows: {len(kaggle_computed_table)}")
print(f"Seasons: {sorted(kaggle_computed_table['Season'].unique())}")
print(f"Teams per season: {kaggle_computed_table.groupby('Season').size().unique()}")
```
